```python
import numpy as np
import pandas as pd
import json
pd.set_option('display.max_rows', None)
```

```python
file_path = './data/data.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
rules_categories_path = 'rulesCategory.json'
with open(rules_categories_path, 'r') as f:
    rule_category = json.load(f)
category_rules = {}

for idx, r_c in enumerate(rule_category):
    category = r_c['type']
    rules = r_c['rules']
    category_rules[category] = rules
    print(f'{idx} - {category} - totle rules: {len(rules)}')
    print(f'rules: {rules}')
```

```python
categories = list(df.columns.values)[:7]
df[categories].head()
```

```python
stylistic_issues_rules = category_rules['Stylistic Issues']
df_StylisticIssues = df[stylistic_issues_rules]
df_StylisticIssues.head()
```

# Variables

```python
variables_rules = category_rules['Variables']
df_varialbes = df[variables_rules]
df_varialbes.head()
```

```python
print('Top 5 variables total')
df_varialbes.sum().sort_values(ascending=False)[:5]
```

```python
df_varialbes.apply(lambda col: col[col != 0].count()).sort_values(ascending=False)[:5]
```

#### 1 no-undef: Any reference to an undeclared variable causes a warning, unless the variable is explicitly mentioned in a /*global ...*/ comment.
#### 2 no-unused-vars: not used var or func (PROBLEM)
#### 3 no-use-before-define: use before define (before ES6 it can)
#### 4 no-shadow-restricted-names: Value Properties of the Global Object (NaN, Infinity, undefined) as well as strict mode restricted identifiers eval and arguments are considered to be restricted names in JavaScript. 
#### 5 no-undef-init: init var to undefined (JS automatically set, so no required to set manually)

```python
no_unused_vars = df_varialbes['no-unused-vars']
has_no_unused_vars = np.count_nonzero(no_unused_vars)
print(f'There are {has_no_unused_vars} of 400 projects ({has_no_unused_vars / 400})')
max_count = np.max(no_unused_vars)
print(f'The one with most are {max_count} in project cstimer')
```

#### 1 no-prototype-builtins: In ECMAScript 5.1, Object.create was added, which enables the creation of objects with a specified [[Prototype]]. #### 2 no-extra-parens: restricts the use of parentheses to only where they are necessary
#### 3 no-cond-assign: mistype a comparison operator e.g.== to =
#### 4 no-empty:Empty block statements
